```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", error_bad_lines=False)

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years

df.head()
```

```python
df['publisher'].value_counts()
```

With your dataframe set up, time to start creating visualizations!

```python
df.sort_values(by=['year'], ascending = True)
```

```python
#make a list with just the publisher and the count of number of books.
#be sure to adjust index and then rename
df_publisher = df['publisher'].value_counts().reset_index().rename(columns = {"index": "publisher", "publisher": "count"}).iloc[:20].set_index('publisher')

df_publisher
```

```python
# Use your Python skills to create a visualization here. matplotlib version
style.use('ggplot')
top_publishers = pd.DataFrame(df_publisher.nlargest(20, 'count')['count'].sort_values(ascending = True))

top_publishers.plot.barh(legend = False, color = 'blue', edgecolor = 'black')
plt.title('Top 20 Publishers from 1900 to 2020', color = 'black')
plt.xlabel('Number of Books Published')
plt.ylabel('Publisher')
plt.xticks(color = 'black')
plt.yticks(color = 'black')

plt.show()
```
